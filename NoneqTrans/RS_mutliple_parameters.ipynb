```python
for k in range(9):
    globals()["width" + str(k+1)] = np.zeros(globals()["nn" + str(k+1)])
    E = 0.1
    for i in range(globals()["nn" + str(k+1)]):
        globals()["phi_" + str(i)] = np.loadtxt(globals()["path" + str(k+1)] + "phi_val_"+ format(E, '.2f')+ "_" +str(1023)+".dat")    
        mx = np.shape(globals()["phi_" + str(i)])[0]
        my = np.shape(globals()["phi_" + str(i)])[1]
        ww = 0
        for j in range(my):
            phi2 = globals()["phi_" + str(i)][:,j]**2
            ww += np.count_nonzero(phi2 < 0.5)
        globals()["width" + str(k+1)][i] = ww/my    
        E += 0.1
```
